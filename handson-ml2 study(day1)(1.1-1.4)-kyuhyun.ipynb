용어 정리

훈련 데이터=훈련 사례(샘플)

성능 측정=정확도( 분류작업에 자주 사용)

### 왜 머신러닝을 사용하는가
1. 기존 프로그래밍 기법-> 패턴을 감지하는 알고리즘 작성 및 프로그램의 패턴 발견시 스팸으로 분류
머신러닝 기법-> 일반 메일에 비해 스팸이 자주 나타나는 패턴 감지및 어떤 단어와 구절이 스펨 메일 판단느데 더 좋은 기준인지 스스로 학습
(코드 간단, 유지보수 쉬움, 정확도 향상)
2. 너무 복잡하거나 알려진 알고리즘 없음(ex.음성인식)
3. 머신러닝 알고리즘이 학습한 것 조사가능
머신러닝 기술 적용해 대용량의 데이터를 분석하며 보이지 않던 패턴 발견 가능(=데이터 마이닝)
4. 유동적 환경(머신러닝 시스템은 새로운 데이터에 적응 가능)

### 애플리케이션 사례
1.생산 라인에서 제품 이미지를 분석해 자동으로 분류<br>
2.뇌를 스캔해 종양 진단<br>
3.자동으로 뉴스기사 분류<br>
4.토론 포럼에서 부정적 코멘트 자동으로 구분<br>
5.긴문서 자동으로 요약<br>
6.챗봇 또는 비서 만들기<br>
7.다양한 성능 지표를 통해 회사의 내년도 수익률 예측 <br>
8.음성 명령에 반응하는 앱 만들기<br>
9.신용카드 부정거래 감지<br>
10.구매 이력을 기반으로 고객을 나누고 각 집합마다 다른 마케팅 전략을 계힉하기<br>
11.고차원의 복잡한 데이터셋을 명확하고 의미 있는 그래프로 표현<br>
12.과거 구매 이력을 기반으로 고객이 관심을 가질 수 있는 상품 추천하기<br>
13.지능형 게임 봇 만들기

### 머신러닝 시스템의 종류
1.사람의 감독하에 훈련하는것인지 아닌것인지(지도, 비지도, 준지도, 강화)<br>
2.실시간으로 점진적인 학습인지 아닌지(온라인, 배치)<br>
3.단순하게 알고 있는 데이터 포인트와 새 데이터 포인트를 비교하는 것인지 아니면 과학자가 하는것처럼 패턴발견후 예측 모형 만드는 것인지

#### 지도 학습과 비지도 학습
지도학습: 알고리즘 주입하는 훈련 데이터에 레이블 추가<br>
-분류: 전형적인 지도 학습 작업(ex.스팸분류)<br>
-회귀: 예측변수라 부르는 특성 사용해 타깃 수치 예측<br>
-로지스틱 회귀: 클래스에 속할 확률 출력<br>
가장 중요한 지도 학습 알고리즘: k-최근접 이웃, 선형 회귀, 로지스틱 회귀,서포트 벡터 머신(SVM), 결정트리와 랜덤 포레스트, 신경망
비지도 학습: 훈련 데이터에 레이블 없음 (아무 도움없이 학습)
-군집: k-평균, DBSCAN, 계층 군집 분석(HCA),
